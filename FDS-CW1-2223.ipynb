## Question 1.2 (10 points)

Write the code in the cell below to implement the changes you listed in Q1.1 so someone presented with it could easily interpret the improved plot. Afterwards, run the code to produce the plot itself. Remember that when you export this notebook to a PDF, it will have a text width of 6 inches.

````python
# Your code for Q1.2 goes here
# Formatting the date column such that we exclude the time. 
df_BTC["Date"] = pd.to_datetime(df_BTC["Date"]).dt.date
df_ETH["Date"] = pd.to_datetime(df_ETH["Date"]).dt.date
df_ADA["Date"] = pd.to_datetime(df_ADA["Date"]).dt.date
# Converting the price data to log values to plot the significant changes in the closing prices
df_BTC["log_close"] = np.log(df_BTC["Close"])
df_ETH["log_close"] = np.log(df_ETH["Close"])
df_ADA["log_close"] = np.log(df_ADA["Close"])

# Plotting the data
fig, ax1 = plt.subplots(1) 
ax1.plot(df_ADA["Date"], df_ADA["log_close"], label = "Cardano",linewidth = 0.7)
ax1.plot(df_ETH["Date"], df_ETH["log_close"], label = "Ethereum",linewidth = 0.7 )
ax1.plot(df_BTC["Date"], df_BTC["log_close"], label = "Bitcoin",linewidth = 0.7)
ax1.set(ylabel="Log value of closing price (USD)",xlabel="Year")
ax1.set_title("Closing price of cryptocurrencies over time (USD)", fontweight = "bold",size= 10,pad = 10)
ax1.legend()
ax1.grid(True,color= "#D3D3D3")
````

Associate countries in the World Happiness Report 2017 and 2018 data sets with a continent, and in which every row has a continent associated with it. You can either
1. Create two DataFrames with continents, one for the 2017 data set and one for the 2018 data set OR

1. Print how many countries do not have a continent associated with them in the new DataFrame
2. Use your extended DataFrame(s) to calculate and print (to 3 decimal places) mean `Happiness score` for each `Continent` for years 2017 and 2018 separately.

````python
# Your code for Q4.2 goes here
tidy_with_continents = tidy_whr
tidy_with_continents = tidy_with_continents.merge(country_continent, on='Country',how='left')

print("Number of Countries that do not have a continent associated with them in the new DataFrame: ",int(tidy_with_continents['Continent'].isna().sum()/2))
Y_17=[]
Y_18=[]
# Mean Happiness Scores(Continent,Year wise):
def mean_continent(yr,cont):
    mean_val = tidy_with_continents[tidy_with_continents.Continent == cont][tidy_with_continents.Year == yr]['Happiness score'].mean()
    return(round(mean_val,4)) # Printing the mean scores upto 3 decimal places
    
continent_list = ['Africa', 'Asia','Europe','North America', 'Oceania','South America' ]
print( "Mean Happiness scores for year 2017: ")
for c in continent_list:
    val = mean_continent(2017,c)
    Y_17.append(val)
    print (c,":",round(val,3))
print("\n")
print( "Mean Happiness scores for year 2018: ")
for c in continent_list:
    val = mean_continent(2018,c)
    Y_18.append(val)
    print (c,":",round(val,3))
````
